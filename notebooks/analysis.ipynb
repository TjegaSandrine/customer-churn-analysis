```python
# Vérifier le type de variables

df.info()
```

TotalCharges est en object au lieu de float

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors = 'coerce')
```

```python
# Verifier les valeurs manquantes
df.isnull().sum()
```

```python
# Gérer les valeurs manquantes

# supprimer les lignes qui contiennent les valeurs manquantes
# df = df.dropna(axis=0).shape

# remplacer les valeurs manquantes de TotalCharges par la moyenne
df['TotalCharges'].fillna(df['TotalCharges'].mean(), inplace=True)
```
